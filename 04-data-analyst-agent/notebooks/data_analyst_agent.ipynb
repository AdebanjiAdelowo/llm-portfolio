# Creating the Agent
This is the easiest Agent we can create with LangChain, we only need to import the **create_pandas_dataframe_agent**.

We create the agent using `ChatOpenAI` with `gpt-4o-mini` at temperature 0 (deterministic output). The `allow_dangerous_code=True` flag is required by newer LangChain versions since the agent executes Python code against your DataFrame.

```python
from langchain_experimental.agents import create_pandas_dataframe_agent
from langchain_openai import ChatOpenAI
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

agent = create_pandas_dataframe_agent(
    llm,
    document,
    verbose=True,
    allow_dangerous_code=True,
    agent_type="tool-calling",
)
```

## First Question — Describe the dataset

```python
agent.invoke("Analyze this data, and write a brief explanation around 100 words.")
```

```python
document.info()
```

## Second Question — Is temperature forecastable?

```python
agent.invoke("Do you think it is possible to forecast the temperature? Explain your reasoning.")
```

## Third Question — CO2 emissions line chart

```python
agent.invoke("Can you create a line graph containing the annual average CO2 emissions over the years?")
```

## Fourth Question — CO2 emissions for Portugal

```python
agent.invoke("Create a line graph with seaborn containing the annual average CO2 emissions in Portugal over the years.")
```

## Last Question — Temperature forecast

```python
agent.invoke("""Select a forecasting model to forecast the temperature.
Use this model to forecast the average temperature for the next 5 years.
Write the forecasted temperatures in a table.""")
```
